```python
import pandas as pd
#syntaxError : unicode error
# 해결법 \ 대신 \\
df1 = pd.read_csv("C:\\Users\\CJA\\Desktop\\user-tweets.csv", encoding='utf-8')
df1.head()
df1.columns
df1['Text'].head()
txt = df1['Text']
print(len(txt))
```

```python
####텍스트 읽어오기####
##영어 텍스트##

line2 = []
for i in range(len(txt)):
    line2.append(txt[i])
# f.close()
print(line2)
```

```python
### 특수문자 제거하기 + 소문자로 변경 ###

# 막 링크가 있는데, 이걸 하니까 그게 다 깨짐
# 이모지는 잘 제거됌
import re
##영어##
compile = re.compile("\W+")
for i in range(len(line2)):

    a = compile.sub(" ",line2[i])
    line2[i] = a.lower()
print(line2)
```

```python
###불용어 제거하기###
# 문장 단위로 구성되어 있으면 어려움? 

##영어##
import nltk
##우선 사용x##
"""nltk.download('all')
nltk.download('wordnet')
nltk.download('stopwords')"""
##############################
from nltk.corpus import stopwords
stop_word_eng = set(stopwords.words('english'))
line2 = [i for i in line2 if i not in stop_word_eng]
line2
```

```python
nltk.download('wordnet')
```

```python
##영어##
### 표제어 추출 사용 시 ###
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
token = RegexpTokenizer('[\w]+')
result_pre_lem = [token.tokenize(i) for i in line2]
middle_pre_lem= [r for i in result_pre_lem for r in i]
final_lem = [lemmatizer.lemmatize(i) for i in middle_pre_lem if not i in stop_word_eng] # 불용어 제거
print(final_lem)
```

```python
###텍스트에서 많이 나온 단어###
###표제어 추출 시#####
import pandas as pd
##영어##
english = pd.Series(final_lem).value_counts().head(20)
print("English top 20")
english
```

```python
##영어##
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer
ps_stemmer = PorterStemmer()
token = RegexpTokenizer('[\w]+')
result2 = [token.tokenize(i) for i in line2]
middle_result2= [r for i in result2 for r in i]
final_result2 = [ps_stemmer.stem(i) for i in middle_result2 if not i in stop_word_eng] # 불용어 제거
print(final_result2)
```

```python
###텍스트에서 많이 나온 단어###(어간 추출 시)
import pandas as pd
##영어##
english2 = pd.Series(final_result2).value_counts().head(10)
print("English top 10")
english2
```
